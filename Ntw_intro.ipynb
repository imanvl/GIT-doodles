# Network analytics with Python and NetworkX
This tutorial will help you answer questions such as:
- What is the overall structure of the network?
- Which are the important countries, or hubs, in the network?
- What are the subgroups and communities in the network?

The setup of this tutorial borrows heavily from [_Exploring and Analyzing Network Data with Python_](
https://programminghistorian.org/en/lessons/exploring-and-analyzing-network-data-with-python)
 by Ladd, Otis, Warren and Weingart


**Our example: The BIS International Banking Statistics**

The [International Banking Statistics](https://www.bis.org/statistics/about_banking_stats.htm) cover the balance sheets of internationally active banks. The locational statistics provide information about the geographical and currency composition of banks' assets and liabilities, including intragroup business. The consolidated statistics measure banks' country risk exposures on a worldwide consolidated basis. Both data sets are collected under the auspices of the Committee on the Global Financial System (CGFS) and reported to the BIS at a country, rather than individual bank, level.


# First step with NetworkX

```
import networkx as nx
import matplotlib.pyplot as plt
graph = nx.Graph()

graph.add_node(1)
graph.add_nodes_from([2,3])
graph.nodes()
```

```
graph.add_edge(1,2)
graph.add_edges_from([(1,2),(2,3)])
graph.edges()
```

```
graph.graph["day"]="Monday"
graph.graph
```

```

graph.add_node(1, time='5pm')
graph.add_nodes_from([3], time='2pm')
# graph.node[1]
graph.nodes[1]['room'] = 714
graph.nodes(data=True)
```

```
graph.add_edge(1, 2, weight=4.7 )
graph.add_edges_from([(3,4),(4,5)], color='red')
graph.add_edges_from([(1,2,{'color':'blue'}), (2,3,{'weight':8})])
graph[1][2]['weight'] = 4.7
# graph.edges[1][2]['weight'] = 4

DG=nx.DiGraph()
DG.add_weighted_edges_from([(1,2,0.5), (3,1,0.75)])
DG.out_degree(1,weight='weight')
DG.degree(1,weight='weight')
DG.successors(1)
DG.neighbors(1)

```

```
nx.draw(graph, with_labels=True)
nx.draw_random(graph)
nx.draw_circular(graph)
nx.draw_spectral(graph)
plt.show()
```
